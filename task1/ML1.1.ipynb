```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df0 = pd.read_csv('housing2.csv')
print(df0.info())
df0.isnull().sum()
```

```python
df0
```

```python
df = df0.fillna(method='bfill')
df
```

```python
df.isnull().sum()
```

```python
df.corr(numeric_only = True)
```

#Longitude and Latitude:
#____There is a strong negative correlation (-0.92) between longitude and latitude.
#Total Rooms :
#____Total_rooms has strong positive correlations with total_bedrooms, population, and households
#Median House Value :
#____Median_house_value has a strong positive correlation with median_income

```python
sns.scatterplot(x='longitude', y='latitude', data=df, alpha=0.5,color='teal')
plt.title('Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
sns.scatterplot(x='total_rooms', y='total_bedrooms', data=df, alpha=0.5,color='teal')
plt.title('Total Rooms vs Total Bedrooms')
plt.xlabel('Total Rooms')
plt.ylabel('Total Bedrooms')
plt.show()
```

```python
sns.scatterplot(x='population', y='households', data=df, alpha=0.5,color='teal')
plt.title('Population vs Households')
plt.xlabel('Population')
plt.ylabel('Households')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only = True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.scatterplot(x='median_income', y='median_house_value', data=df, alpha=0.5,color='teal')
plt.title('Median Income vs Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```
